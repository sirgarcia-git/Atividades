```python
# Gerando um arquivo COG
import os
import subprocess

def process_file(img_path):
    filename = img_path.name
    output_file = output_dir / filename

    cmd = [
        "gdal_translate",
        "-ot", "Byte",
        "-scale", "100", "2000", "0", "255",
        "-of", "COG",
        "-co", "COMPRES=LZW",
        "-co", "BLOCKSIZE=512",
        "-co", "BIGTIFF=YES",
        str(img_path),
        str(output_file)
    ]

    proc = subprocess.run(cmd, capture_output=True)
    if proc.returncode == 0:
        return (filename, "ok")
    else:
        return (filename, "fail", proc.stderr.decode())

from pathlib import Path

# 1. Define os caminhos das pastas de entrada e saída
diretorio_entrada = Path("/home/jovyan/Downloads/")
output_dir = Path("/home/jovyan/Downloads/COG")

# 2. Varre a pasta e processe cada imagem encontrada com a extensão .tif
for arquivo in diretorio_entrada.glob("*.tif"):
    resultado = process_file(arquivo)
    print(resultado)
```
