```python
# Gerando arquivo COG
import os
import subprocess
from pathlib import Path

arquivo_original = 'https://data.inpe.br/bdc/data/CB4A-MUX-L4-SR/2026_06/CBERS_4A_MUX_RAW_2026_06_19.13_17_01_ETC2/213_113_0/4_BC_UTM_WGS84/CBERS_4A_MUX_20260619_213_113_L4_BAND5_GRID_SURFACE.tif'
arquivo_executado = '/home/jovyan/Downloads/resultado_cog.tif'

def process_file(img_url, output_path):
    # Extrai o nome do arquivo a partir da URL para usar no retorno
    filename = img_url.split("/")[-1]
    
    # Adiciona o prefixo /vsicurl/ para o GDAL ler a URL diretamente da web
    vsi_url = f"/vsicurl/{img_url}"

    cmd = [
        "gdal_translate",
        "-ot", "Byte",
        "-scale", "100", "2000", "0", "255",
        "-of", "COG",
        "-co", "COMPRESS=LZW",
        "-co", "BLOCKSIZE=512",
        "-co", "BIGTIFF=YES",
        vsi_url,                # URL virtualizada
        str(output_path)        # Destino final local
    ]

    proc = subprocess.run(cmd, capture_output=True)
    
    if proc.returncode == 0:
        return (filename, "executado ok")
    else:
        return (filename, "fail", proc.stderr.decode())

# Executa a função e imprime o resultado
resultado = process_file(arquivo_original, arquivo_executado)
print(resultado)
```
